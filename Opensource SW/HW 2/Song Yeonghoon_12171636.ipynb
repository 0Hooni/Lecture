```python
import pandas
import numpy
import math

#데이터 호출
data = pandas.read_csv("ml-100k/u.data", sep="\t", names=['User_ID','Movie_ID', 'Rating','TimeStamp'])
user = pandas.read_csv("ml-100k/u.user", sep="|", names=['User_ID','Age', 'Gender','Occupation', 'Zip Code'])
item = pandas.read_csv("ml-100k/u.item", sep="|",
                       names = ['Movie_ID','movie title','release date',
                                'video release date', 'IMDb URL','unknown',
                                'Action','Adventure','Animation','Children\'s',
                                'Comedy','Crime','Documentary','Drama','Fantasy',
                                'Film-Noir','Horror','Musical','Mystery','Romance',
                                'Sci-Fi','Thriller','War','Western'])
genre = pandas.read_csv("ml-100k/u.genre", sep="|", names=['genre_name', 'genre_code'])
```

```python
#데이터 병합
D_U = pandas.merge(data, user, on='User_ID')
D_U_I = pandas.merge(D_U, item, on='Movie_ID')

#평점 카운팅
rate_CNT = D_U_I.pivot_table('Rating', aggfunc='count', index = 'movie title')
rate_CNT = rate_CNT.sort_values(by='Rating', ascending=False)
pandas.options.display.float_format = '{:.0f}'.format
display(rate_CNT[:100])

#평점 평균
pandas.options.display.float_format = '{:.2f}'.format
rate_AVG = D_U_I.pivot_table('Rating', aggfunc='mean', index = 'movie title')
rate_AVG = rate_AVG.sort_values(by='Rating', ascending=False)
display(rate_AVG[:100])
```

```python
rate_CNT = D_U_I.pivot_table('Rating', aggfunc='count', index = 'movie title', columns ='Gender')
rate_CNT = rate_CNT.sort_values(by='F', ascending=False)

#평점 카운팅
pandas.options.display.float_format = '{:.0f}'.format
print("Sort based on F")
display(rate_CNT[:100])

#평점 평균
rate_AVG = D_U_I.pivot_table('Rating', aggfunc='mean', index = 'movie title', columns ='Gender')
rate_AVG = rate_AVG.sort_values(by='F', ascending=False)
pandas.options.display.float_format = '{:.2f}'.format
print("Sort based on F")
display(rate_AVG[:100])
```

```python
rate_CNT = D_U_I.pivot_table('Rating', aggfunc='count', index = 'movie title', columns ='Gender')
rate_CNT = rate_CNT.sort_values(by='M', ascending=False)
pandas.options.display.float_format = '{:.0f}'.format
print("Sort based on M")
display(rate_CNT[:100])

rate_AVG = D_U_I.pivot_table('Rating', aggfunc='mean', index = 'movie title', columns ='Gender')
rate_AVG = rate_AVG.sort_values(by='M', ascending=False)
pandas.options.display.float_format = '{:.2f}'.format
print("Sort based on M")
display(rate_AVG[:100])
```

```python
#직업별 평점 카운트
Occu_data = pandas.read_csv("ml-100k/u.occupation", sep="\t", names=['Occupation'])
rate_CNT = D_U_I.pivot_table('Rating', aggfunc='count', index = 'movie title', columns ='Occupation')
pandas.options.display.float_format = '{:.0f}'.format

#직업의 종류수 만큼 반복해서 출력
for i in range(0, 21):
    name = Occu_data.iloc[i, 0]
    rate_CNT = rate_CNT.sort_values(by=name, ascending=False)
    print("Sort based on " + name)
    display(rate_CNT[:100])
```

```python
#직업별 평점 평균
rate_AVG = D_U_I.pivot_table('Rating', aggfunc='mean', index = 'movie title', columns ='Occupation')
pandas.options.display.float_format = '{:.2f}'.format
for i in range(0, 21):
    name = Occu_data.iloc[i, 0]
    rate_AVG = rate_AVG.sort_values(by=name, ascending=False)
    print("Sort based on " + name)
    display(rate_AVG[:100])
```

```python
Genre_data = pandas.read_csv("ml-100k/u.genre", sep="|", names=['Genre', 'Genre_code'])


#영화별로 평점 및 장르와 테이블 병합
rate_CNT = D_U_I.pivot_table('Rating', aggfunc='count', index = 'movie title')
rate_CNT = rate_CNT.sort_values(by='Rating', ascending=False)

#데이터를 담을 테이블 초기화
Genre_Table = pandas.DataFrame({'Genre' : ['NULL'], 'Count':[0]})
Count_Frame = pandas.DataFrame(data=Genre_Table)

for i in range(0, 19):
    now_Genre_Name = Genre_data.iloc[i,0]   # 장르 이름
    data_to_insert = {'Genre': [now_Genre_Name], 'Count': [0]}
    df_to_insert = pandas.DataFrame(data=data_to_insert)
    Count_Frame = Count_Frame.append(df_to_insert, ignore_index=True)
    
Count_Frame = Count_Frame.drop(0)

rate_CNT = D_U_I.pivot_table('Rating', aggfunc='count', index = 'movie title')
C_I = pandas.merge(rate_CNT, item, on='movie title')

for j in range(6, 6+len(Count_Frame)) : 
    for i in range(0, len(C_I)):
        if C_I.iloc[i, j] == 1 :
            now_CNT = C_I.iloc[i, 1]
            Count_Frame.iloc[j-6, 1] += now_CNT #평점 누적

pandas.options.display.float_format = '{:.0f}'.format
rate_CNT_Genre = Count_Frame.sort_values(by='Count', ascending=False)
display(rate_CNT_Genre)
```

```python
# 데이터를 담을 테이블 초기화
Genre_Table = pandas.DataFrame({'Genre' : ['NULL'], 'M_Count': [0], 'F_Count': [0]})
Count_Frame = pandas.DataFrame(data=Genre_Table)

for i in range(0, 19):
    now_Genre_Name = Genre_data.iloc[i,0]   # 장르 이름
    data_to_insert = {'Genre': [now_Genre_Name], 'M_Count': [0], 'F_Count': [0]}
    df_to_insert = pandas.DataFrame(data=data_to_insert)
    Count_Frame = Count_Frame.append(df_to_insert, ignore_index=True)
    
Count_Frame = Count_Frame.drop(0)

rate_CNT = D_U_I.pivot_table('Rating', aggfunc='count', index = 'movie title', columns ='Gender')
C_I = pandas.merge(rate_CNT, item, on='movie title')


for j in range(7, 7+len(Count_Frame)) : 
    for i in range(0, len(C_I)):
        if C_I.iloc[i, j] == 1 :
            now_CNT_F = C_I.iloc[i, 1]
            now_CNT_M = C_I.iloc[i, 2]
            if not math.isnan(now_CNT_F):
                Count_Frame.iloc[j-7, 1] += now_CNT_F #평점 누적
            if not math.isnan(now_CNT_M):
                Count_Frame.iloc[j-7, 2] += now_CNT_M #평점 누적

pandas.options.display.float_format = '{:.0f}'.format
rate_CNT_Genre_Gender = Count_Frame.sort_values(by='M_Count', ascending=False)
print("Sort based on M")
display(rate_CNT_Genre_Gender)

print("Sort based on F")
rate_CNT_Genre_Gender = Count_Frame.sort_values(by='F_Count', ascending=False)
display(rate_CNT_Genre_Gender)
```
